```python
import requests
```

```python
base_site = 'https://en.wikipedia.org/wiki/Music'
```

```python
response = requests.get(base_site)
```

```python
response.ok
content = response.text
```

```python
from bs4 import BeautifulSoup
```

```python
read = BeautifulSoup(content)
```

```python
links = read.find_all("a")
```

```python
relativ_url = links[1] ["href"]
relativ_url
```

```python
relative_urlss = [i.get("href")for i in  links if i.get("href") != None]
```

```python
relative_urlss
```

```python
from urllib.parse import urljoin
```

```python
full_urls = [urljoin(base_site,url)for url in relative_urlss]
```

```python
for i in full_urls:
    print(i)
```

```python
internal_url = [url for url in full_urls if "wikipedia.org" in url]
```
